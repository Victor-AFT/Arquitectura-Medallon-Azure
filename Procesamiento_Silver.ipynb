```python
from pyspark.sql.functions import col, unix_timestamp, struct, to_json, collect_list

# -------------------------------------------------
# CONFIG STORAGE
# -------------------------------------------------

spark.conf.set(
"fs.azure.account.key.XXXX.blob.core.windows.net",
"KEY_XXXX"
)

base = "wasbs://silver@XXXX.blob.core.windows.net/"

# -------------------------------------------------
# LOAD TABLES
# -------------------------------------------------

aircraft = spark.read.csv(base+"aircraft.csv", header=True, inferSchema=True)
airline = spark.read.csv(base+"airline.csv", header=True, inferSchema=True)
flights = spark.read.csv(base+"flights.csv", header=True, inferSchema=True)
flight_numbers = spark.read.csv(base+"flight_info.csv", header=True, inferSchema=True)
departure = spark.read.csv(base+"departure.csv", header=True, inferSchema=True)
arrival = spark.read.csv(base+"arrival.csv", header=True, inferSchema=True)

aircraft.display()
airline.display()
flights.display()
flight_numbers.display()
departure.display()
arrival.display()

```

```python
# -------------------------
# DIM AIRLINE
# -------------------------
dim_airline = airline.dropDuplicates(["iata"]).na.fill("null")

# -------------------------
# DIM AIRCRAFT
# -------------------------
dim_aircraft = aircraft.dropDuplicates(["registration"]).na.fill("null")

# -------------------------
# DIM FLIGHT
# -------------------------
dim_flight = flights.dropDuplicates(["flight_id"]).na.fill("null")

# -------------------------
# DIM FLIGHT NUMBER
# -------------------------
dim_flight_number = flight_numbers.dropDuplicates(["flight_id"]).na.fill("null")

# -------------------------
# DIM AIRPORT
# -------------------------
dep_airport = departure.select(
    col("airport").alias("airport"),
    col("iata"),
    col("icao"),
    col("timezone")
).na.fill("null")

arr_airport = arrival.select(
    col("airport").alias("airport"),
    col("iata"),
    col("icao"),
    col("timezone")
).na.fill("null")

#dim_airport = dep_airport.union(arr_airport).dropDuplicates(["airport"])
```

```python
fact_flights = (
    flights.alias("f")
    .join(airline.alias("al"), "flight_id", "left")
    .join(aircraft.alias("ac"), "flight_id", "left")
    .join(departure.alias("d"), "flight_id", "left")
    .join(arrival.alias("a"), "flight_id", "left")
    .join(flight_numbers.alias("fn"), "flight_id", "left")
    .select(
        col("f.flight_id"),
        col("f.flight_date"),
        col("f.flight_status"),

        col("al.iata").alias("airline_iata"),
        col("ac.registration").alias("aircraft_registration"),

        col("d.airport").alias("departure_airport"),
        col("a.airport").alias("arrival_airport"),

        col("d.scheduled").alias("scheduled_departure"),
        col("d.actual").alias("actual_departure"),
        col("d.delay").alias("departure_delay"),

        col("a.scheduled").alias("scheduled_arrival"),
        col("a.actual").alias("actual_arrival"),
        col("a.delay").alias("arrival_delay"),

        col("d.gate").alias("departure_gate"),
        col("a.gate").alias("arrival_gate"),
        col("d.terminal").alias("departure_terminal"),
        col("a.terminal").alias("arrival_terminal")
    )
)
```

```python
json_unificado = (
    flights.alias("f")
    .join(airline.alias("al"), "flight_id", "left")
    .join(aircraft.alias("ac"), "flight_id", "left")
    .join(departure.alias("d"), "flight_id", "left")
    .join(arrival.alias("a"), "flight_id", "left")
    .join(flight_numbers.alias("fn"), "flight_id", "left")
    .select(
        col("f.flight_id"),
        col("f.flight_date"),
        col("f.flight_status"),

        struct(
            col("fn.flight_id"),
            col("fn.number"),
            col("fn.iata"),
            col("fn.icao")
        ).alias("flight_number"),

        struct(
            col("al.flight_id"),
            col("al.name"),
            col("al.iata"),
            col("al.icao")
        ).alias("airline"),

        struct(
            col("ac.flight_id"),
            col("ac.registration"),
            col("ac.iata"),
            col("ac.icao"),
            col("ac.icao24")
        ).alias("aircraft"),

        struct(
            col("d.flight_id"),
            col("d.airport"),
            col("d.scheduled"),
            col("d.actual"),
            col("d.delay"),
            col("d.gate"),
            col("d.terminal"),
            col("d.estimated_runway")
        ).alias("departure"),

        struct(
            col("a.flight_id"),
            col("a.airport"),
            col("a.scheduled"),
            col("a.actual"),
            col("a.delay"),
            col("a.gate"),
            col("a.terminal"),
            col("a.estimated_runway")
        ).alias("arrival")
    )
)
```

```python

base = "wasbs://gold@XXXXwindows.net/"
temp_output = base + "_tmp_json/"
final_file = base + "vuelos.json"
# -------------------------------------------------
# WRITE TEMP JSON (1 FILE)
# -------------------------------------------------
(
    json_unificado
    .coalesce(1)
    .write
    .mode("overwrite")
    .json(temp_output)
)

# -------------------------------------------------
# RENAME PART → vuelos.json
# (solo funciona en Databricks / Synapse / Fabric)
# -------------------------------------------------
files = dbutils.fs.ls(temp_output)
json_file = [f.path for f in files if f.path.endswith(".json")][0]

dbutils.fs.mv(json_file, final_file, True)
dbutils.fs.rm(temp_output, True)

print("✅ Archivo final creado:", final_file)
```
